**Netflix**! What started in 1997 as a DVD rental service has since exploded into one of the largest entertainment and media companies.

Given the large number of movies and series available on the platform, it is a perfect opportunity to flex your exploratory data analysis skills and dive into the entertainment industry. Our friend has also been brushing up on their Python skills and has taken a first crack at a CSV file containing Netflix data. They believe that the average duration of movies has been declining. Using your friends initial research, you'll delve into the Netflix data to see if you can determine whether movie lengths are actually getting shorter and explain some of the contributing factors, if any.

You have been supplied with the dataset `netflix_data.csv` , along with the following table detailing the column names and descriptions:

## The data
### **netflix_data.csv**
| Column | Description |
|--------|-------------|
| `show_id` | The ID of the show |
| `type` | Type of show |
| `title` | Title of the show |
| `director` | Director of the show |
| `cast` | Cast of the show |
| `country` | Country of origin |
| `date_added` | Date added to Netflix |
| `release_year` | Year of Netflix release |
| `duration` | Duration of the show in minutes |
| `description` | Description of the show |
| `genre` | Show genre |

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# loading the csv file into a dataframe named as netflix_df
netlix_df = pd.read_csv('netflix_data.csv')
# filtering the dataset where type is not TV show and storing in netflix_subset
is_not_TV_show = netlix_df['type'] != 'TV Show'
netflix_subset = netlix_df[is_not_TV_show]

# subsetting the dataframe for desired columns
required_columns = ['title', 'country', 'genre', 'release_year', 'duration']
netflix_movies = netflix_subset[required_columns]

# filtering data where movies are shorter than 60 minutes
is_shorter_than_60 = netflix_movies['duration'] < 60
short_movies = netflix_movies[is_shorter_than_60]

# assigning different colours to the generes
colors = []

for genre in netflix_movies['genre']:
    if genre == 'Children':
        colors.append('green')
    elif genre == 'Documentaries':    
        colors.append('yellow')
    elif genre == 'Stand-Up':    
        colors.append('blue')
    else:    
        colors.append('red')   

# initializing matplotlib figure object
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

# scatter plot for release year and duration of the movies
plt.scatter(x = netflix_movies['release_year'], y = netflix_movies['duration'], c = colors, s=1.5)
axes.set_xlabel('Release year')
axes.set_ylabel('Duration (min)')
axes.set_title('Movie Duration by Year of Release')
plt.show()

# are we certain that the movies are getting shorter by the passage of time?
answer = "no"
```
